```python
t_cnt = 0
for i in range(23,24):
    # 책 카테고리 
    category = driver.find_element(By.XPATH,f'/html/body/div/div[2]/div[1]/div[1]/ul/li[1]/ul/li[{i+1}]/a') 
    # print(category.text)
    # yes24['category'].append(category.text)
    category_1 = category.text
    category.click()

    time.sleep(1)

    # 20개씩 보기에서 80개씩 보기 위해 변경
    view_20 = driver.find_element(By.XPATH,'//*[@id="bestList"]/div[3]/div[2]/div[1]/select[1]') 
    view_20.click() 

    # 80개씩 보기
    view_80 = driver.find_element(By.XPATH,'//*[@id="bestList"]/div[3]/div[2]/div[1]/select[1]/option[3]') 
    view_80.click() 

    for q in range(3, 12):
        for j in range(1, 160, 2):
            try : # 예외 처리
                # 예약 판매 중이거나 리뷰 개수가 하나도 없으면 오류 발생
                # 오류 제어
                review_number = driver.find_element(By.XPATH,f'//*[@id="category_layout"]/tbody/tr[{j}]/td[3]/p[3]/a')
                review_number = review_number.text
                review_number = int(review_number[:-1])

                # 리뷰 개수가 10개 이상일 때
                if review_number >= 10:
                    breaker = False
                    # 책 제목
                    title = driver.find_element(By.XPATH,f'//*[@id="category_layout"]/tbody/tr[{j}]/td[3]/p[1]/a[1]')
                    title_1 = title.text
                    print(title_1)
                    t_cnt += 1
                    # yes24['title'].append(title.text)
                    if t_cnt < 248:
                        continue
                    # 가격
                    price = driver.find_element(By.XPATH,f'//*[@id="category_layout"]/tbody/tr[{j}]/td[3]/p[2]')
                    price = price.text.split('원')[0]
                    price = price.replace(',','')
                    # yes24['price'].append(price)
                    # print(price)

                    title.click()

                    time.sleep(1)

                    # 책 저자
                    author = driver.find_element(By.XPATH,f'//*[@id="yDetailTopWrap"]/div[2]/div[1]/span[2]/span[1]/a')
                    # yes24['author'].append(author.text)
                    # print(author.text)
                    author = author.text

                    # 책 출간일
                    publication_date = driver.find_element(By.XPATH,f'//*[@id="yDetailTopWrap"]/div[2]/div[1]/span[2]/span[3]')
                    # yes24['publication_date'].append(publication_date.text)
                    # print(publication_date.text)
                    publication_date = publication_date.text

                    # 책 링크
                    book_link = driver.current_url
                    # yes24['book_link'].append(book_link)

                    # 책 전체 평점
                    book_grade = driver.find_element(By.XPATH,f'//*[@id="spanGdRating"]/a/em')
                    book_grade = float(book_grade.text)
                    # yes24['book_grade'].append(book_grade)
                    # print(book_grade)

                    # 세부 평점 보기
                    button = driver.find_element(By.XPATH,'//*[@id="spanGdRating"]/span/span[1]/a') 
                    button.click()

                    # 책 전체 평점 : 내용
                    book_grade_1 = driver.find_element(By.XPATH,f'//*[@id="spanGdRating"]/span/span[2]/span/dl[1]/dd/span')
                    book_grade_1 = float(book_grade_1.text[:-1])
                    # yes24['book_grade_1'].append(book_grade_1)
                    # print(book_grade_1)

                    # 책 전체 평점 : 편집/디자인
                    book_grade_2 = driver.find_element(By.XPATH,f'//*[@id="spanGdRating"]/span/span[2]/span/dl[2]/dd/span')
                    book_grade_2 = float(book_grade_2.text[:-1])
                    # yes24['book_grade_2'].append(book_grade_2)
                    # print(book_grade_2)

                    # 리뷰 개수 
                    # review_number != num 다를 때 있음
                    num = driver.find_element(By.XPATH,f'//*[@id="emReviewCountText"]')
                    num = int(num.text[1:-1])
                    # print('리뷰 개수 : ',num)

                    cnt = 0
                    while True:
                        # 다음 칸 으로 넘기기
                        for a in range(3, 13):
                            # 한 페이지 당 리뷰 갯수
                            for k in range(2,7):
                                yes24['category'].append(category_1)
                                yes24['title'].append(title_1)
                                yes24['price'].append(price)
                                yes24['book_link'].append(book_link)
                                yes24['author'].append(author)
                                yes24['publication_date'].append(publication_date)
                                yes24['book_grade'].append(book_grade)
                                yes24['book_grade_1'].append(book_grade_1)
                                yes24['book_grade_2'].append(book_grade_2)

                                # print(a, ' ', k)
                                # 내용 별점
                                grade_1 = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[{k}]/div[1]/div/span/span[1]')
                                grade_1 = float(grade_1.text[2:-1])
                                yes24['grade_1'].append(grade_1)
                                # grade_1

                                # 편집/디자인 별점
                                grade_2 = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[{k}]/div[1]/div/span/span[2]')
                                grade_2 = float(grade_2.text[2:-1])
                                yes24['grade_2'].append(grade_2)
                                # grade_2

                                is_buy = 0
                                byy = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[{k}]/div[1]')
                                if '구매' in byy.text:
                                    is_buy = 1
                                yes24['is_buy'].append(is_buy)
                                # print('구매햇념? ', is_buy)

                                # 더보기 버튼 클릭
                                more = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[{k}]/div[2]/a/div/span')
                                more.click()

                                time.sleep(1)

                                # 리뷰 텍스트 긁어오기
                                review_text = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[{k}]/div[3]/div[2]')
                                review_text = review_text.text
                                yes24['review'].append(review_text)
                                # print(review_text.text)
                                cnt += 1
                                # print(cnt)
                                time.sleep(1)

                                # user_id 가져오기
                                user = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[{k}]/div[1]/div/em[2]/a')

                                # 아예 블로그도 닉네임도 안적혀있네,,,
                                if bool(user.text):
                                    user.click()
                                    # print(1)
                                    go_blog = driver.find_element(By.XPATH,f'//*[@id="blogMnuTTip"]/ul/li[1]/a')
                                    go_blog.click()
                                    # print(2)


                                    # 최근 열립 탭으로 전환
                                    driver.switch_to.window(driver.window_handles[-1])
                                    # print(3)
                                    time.sleep(1)

                                    try:
                                        # WebDriverWait(driver, 3).until(EC.alert_is_present(),'이 회원님은 블로그를 개설하지 않으셨습니다.')
                                        alert = driver.switch_to.alert

                                        # 취소하기(닫기)
                                        alert.dismiss()

                                        # 확인하기
                                        alert.accept()
                                    except : 
                                        pass

                                    time.sleep(1)

                                    if len(driver.window_handles) == 1:
                                        print('!')
                                        driver.switch_to.window(driver.window_handles[0])

                                    # 현재 url
                                    user_link = driver.current_url
                                    # print(user_link)
                                    # print(4)

                                    if user_link == None:
                                        user_id = None
                                    else:
                                        user_id = user_link.split('=')[-1]

                                        time.sleep(3)

                                        if len(driver.window_handles) == 1:
                                            pass
                                        else:
                                            # 현재 탭 닫기
                                            driver.close()
                                    # print(user_id)
                                    # 맨 처음 탭으로 변경(0번 탭)
                                    driver.switch_to.window(driver.window_handles[0])

                                    yes24['user_id'].append(user_id)
                                else:
                                    user_id = None
                                    yes24['user_id'].append(user_id)

                                # print(user_id)
                                time.sleep(2)

                                if cnt == num:
                                    print(cnt)
                                    breaker = True
                                    break
                            if breaker == True:
                                break
                            if a == 12:
                                pass
                            else:
                                naxt_page = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[7]/div[1]/div/a[{a}]')
                                naxt_page.click()
                                time.sleep(2)
                        if breaker == True:
                            break

                        naxt_page = driver.find_element(By.XPATH,f'//*[@id="infoset_reviewContentList"]/div[7]/div[1]/div/a[12]')
                        naxt_page.click()
                        time.sleep(2)
                    driver.back()
            except:
                pass
        next_80_page = driver.find_element(By.XPATH,f'//*[@id="bestList"]/div[3]/div[1]/div[1]/p/a[{q}]')
        next_80_page.click()
        time.sleep(2)
```
